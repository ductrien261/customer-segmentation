**Xử lý dữ liệu**

```python
data = df.select_dtypes(include=[np.number]).drop(columns=['Customer Id'])

Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
data = data[~((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))).any(axis=1)]
print(f"Số lượng dữ liệu sau khi loại ngoại lệ: {len(data)}")
print("\nMô tả dữ liệu sau khi loại ngoại lệ:")
data.describe()
```

```python
# Chuẩn hóa dữ liệu
def normalize(X):
    X_norm = (X - X.mean()) / X.std()
    return X_norm

X = normalize(data[['Age', 'Income', 'DebtIncomeRatio', 'Years Employed']])

print("\nDữ liệu sau khi chuẩn hóa:")
X.describe().style.format("{:.6f}")
```

**Phân cụm**

```python
# Hàm tính khoảng cách Euclidean
def euclidean(a, b):
    return np.sqrt(np.sum((a - b) ** 2))

# Khởi tạo centroid bằng K-means++
def init_centroids_kmeans(X, k):
    X_np = X.to_numpy()
    centroids = [X_np[np.random.randint(len(X_np))]]
    for _ in range(1, k):
        distances = np.array([min([euclidean(x, c) for c in centroids]) for x in X_np])
        probs = distances / distances.sum()
        cumprobs = probs.cumsum() #Tính xác suất tích luỹ
        r = np.random.random()
        for j, p in enumerate(cumprobs):
            if r < p:
                centroids.append(X_np[j])
                break
    return np.array(centroids)

# Gán điểm dữ liệu vào cụm gần nhất
def assign_clusters(X, centroids):
    clusters = []
    for x in X.to_numpy():
        distances = [euclidean(x, c) for c in centroids]
        cluster = np.argmin(distances)
        clusters.append(cluster)
    return clusters

# Cập nhật centroid mới
def update_centroids(X, clusters, k):
    new_centroids = []
    X_np = X.to_numpy()
    for i in range(k):
        points = X_np[np.array(clusters) == i]
        if len(points) == 0:
            # Nếu cụm rỗng, chọn lại ngẫu nhiên
            new_centroids.append(X_np[np.random.randint(len(X_np))])
        else:
            new_centroids.append(np.mean(points, axis=0))
    return new_centroids

# Hàm K-means chính
def kmeans(X, k, max_iter=100, tol=1e-4):
    centroids = init_centroids_kmeans(X, k)
    for i in range(max_iter):
        clusters = assign_clusters(X, centroids)
        new_centroids = update_centroids(X, clusters, k)
        shifts = [euclidean(a, b) for a, b in zip(centroids, new_centroids)]
        if max(shifts) < tol:
            print(f"Hội tụ sau {i+1} vòng lặp")
            break
        centroids = new_centroids
    return clusters, centroids
```

```python
k = 4
clusters, centroids = kmeans(X, k)

X['Cluster'] = clusters
print("\nSố điểm trong từng cụm:")
print(X['Cluster'].value_counts())
```

**Kết quả**

```python
colors = ['red', 'green', 'blue', 'orange', 'purple', 'yellow', 'pink']

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 5))

# Biểu đồ 1: Age vs Income
for i in range(k):
    ax1.scatter(X[X['Cluster'] == i]['Income'], X[X['Cluster'] == i]['Age'],
                label=f'Cluster {i}', alpha=0.6, color=colors[i])
for c in centroids:
    ax1.scatter(c[1], c[0], marker='x', color='black', s=100, linewidths=3)  # c[1]: Income, c[0]: Age
ax1.set_xlabel("Income (normalized)")
ax1.set_ylabel("Age (normalized)")
ax1.set_title("Age vs Income")
ax1.legend()
ax1.grid(True)

# Biểu đồ 2: Age vs DebtIncomeRatio
for i in range(k):
    ax2.scatter(X[X['Cluster'] == i]['DebtIncomeRatio'], X[X['Cluster'] == i]['Age'],
                label=f'Cluster {i}', alpha=0.6, color=colors[i])
for c in centroids:
    ax2.scatter(c[2], c[0], marker='x', color='black', s=100, linewidths=3)  # c[2]: DebtIncomeRatio, c[0]: Age
ax2.set_xlabel("DebtIncomeRatio (normalized)")
ax2.set_ylabel("Age (normalized)")
ax2.set_title("Age vs DebtIncomeRatio")
ax2.legend()
ax2.grid(True)

# Biểu đồ 3: Age vs Years Employed
for i in range(k):
    ax3.scatter(X[X['Cluster'] == i]['Years Employed'], X[X['Cluster'] == i]['Age'],
                label=f'Cluster {i}', alpha=0.6, color=colors[i])
for c in centroids:
    ax3.scatter(c[3], c[0], marker='x', color='black', s=100, linewidths=3)  # c[3]: Years Employed, c[0]: Age
ax3.set_xlabel("Years Employed (normalized)")
ax3.set_ylabel("Age (normalized)")
ax3.set_title("Age vs Years Employed")
ax3.legend()
ax3.grid(True)

plt.tight_layout()
plt.show()
```
